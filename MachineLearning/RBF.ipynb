```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def dsigmoid(x):
    return (1 - sigmoid(x)) * sigmoid(x)

def dloss(t, out):
    return (out - t) / (out * (1 - out))
```

```python
w1 = 1
w2 = 1
b = 1

X = [[0.88, 0.53], [0.88, 0.53], [0.53, 0.88], [0.53, 0.88]]
Y = [1, 0, 0, 1]
```

```python
def update(x, t):
    global w1
    global w2
    global b
    p = w1 * x[0] + w2 * x[1] + b
    out = sigmoid(p)
    print(f'p = {w1} * {x[0]} + {w2} * {x[1]} + {b} = {p} => out = {out}')
    
    dLdw1 = dloss(t, out) * dsigmoid(p) * x[0]
    dLdw2 = dloss(t, out) * dsigmoid(p) * x[1]
    dLdb = dloss(t, out) * dsigmoid(p)
    w1 -= dLdw1
    w2 -= dLdw2
    b -= dLdb
    
    print(f'dL/dw1 = {dLdw1} => w1 = {w1}')
    print(f'dL/dw2 = {dLdw2} => w2 = {w2}')
    print(f'dL/db = {dLdb} => b = {b}')
```

```python
for epoch in range(0, 3):
    print('------------------------------------------------')
    print(f'----------------- EPOCH {epoch + 1} -------------')
    for i, (xs, ys) in enumerate(zip(X, Y)):
        print(f'Step: {i + 1}')
        update(xs, ys)
```
